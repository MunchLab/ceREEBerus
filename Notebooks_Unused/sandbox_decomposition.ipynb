```python
from cereeberus import ReebGraph 

import cereeberus.data.ex_reebgraphs as ex_rg
import numpy as np
```

```python
R = ex_rg.dancing_man()
R.draw()
```

```python
# Get a path that continues upward in the reeb grpah until it hits a local max 

def upward_path(R, v): 
    path = [v]
    while R.up_degree(path[-1]) > 0: 
        s = next(R.successors(path[-1]))
        path.append(s)
            
            
    return path

v = 7 # I know this is a local min already
path = upward_path(R, v)
print(path)
```

```python
R.remove_path_from(path)
```

```python
R.draw()
```

```python
R.sorted_vertices()[0]
```

```python
branches = [] 

for v in R.sorted_vertices(): 
    print(f"Vertex: {v}, f(v) = {R.f[v]}, degree = {R.down_degree(v)}")
    
    # get a maximal path upward from v 
    path = upward_path(R,v)
    
    start_f = R.f[path[0]]
    end_f = R.f[path[-1]]
    branches.append((start_f, end_f, len(branches), len(branches)))
    
branches = np.array(branches)
print(branches)
```
