Generate Data

```python
from sklearn.datasets import make_circles

number_of_points = 300

data, labels = make_circles(n_samples=number_of_points, factor=0.5, noise=0.03, random_state=0)


val = 0
pointcloud = []
while val < number_of_points:
    pointcloud.append(data[val])
    val += 1
```

```python
import matplotlib.pyplot as plt

plt.scatter(data[:, 0], data[:, 1], c=labels)
plt.axis('scaled');
```

Simple Mapper (no parameters except input graph)

```python
from cereeberus import MapperGraph
import networkx as nx
import sklearn

graph = MapperGraph()
graph = MapperGraph.runmapper(pointcloud, "y", MapperGraph.cover(-1.1,.5,.1,20), sklearn.cluster.DBSCAN(min_samples=2,eps=0.15).fit)
graph.draw()

```

```python
graph2 = MapperGraph.runmapper([(0.6, 0), (-0.1, 0.5)], "x", [(-1,0),(-0.5,0.5),(0,1)], "trivial")
graph2.draw()
```
