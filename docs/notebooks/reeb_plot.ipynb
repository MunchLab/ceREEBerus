# Examples of Reeb graphs

In the following notebook, we show examples of Reeb graphs that are included in the package which show functionality related to the main class. All Reeb graphs are stored as the class `cereberus.reeb.Reeb`.

```python
reeb_torus = ex_reebgraphs.torus()
reeb_torus.draw()
```

The Reeb Graph class has many attributes available such as nodes, edges, position, up degree, down degree, and function values provided at class instantiation

```python
print("Reeb Graph nodes: ", reeb_torus.nodes)
print("Reeb Graph edges: ", reeb_torus.edges)
print("Reeb Graph function values: ", reeb_torus.f)
# print("Reeb Graph up degree: ", reeb_torus.up_deg)
# print("Reeb Graph down degree: ", reeb_torus.down_deg)
#print("Reeb Graph Adjacency Matrix:", reeb_torus.adjacency)
print("Reeb Graph Plotting Positions:", reeb_torus.pos_f)
# print("Drawing Preference:", reeb_torus._horizontalDrawing)
print(reeb_torus)
```

There is also a custom plot function to plot Reeb Graphs which can be called with `plot_reeb`.

```python
reeb_torus.draw()
```

```python
reeb_torus.draw_straight_line(with_labels=True)
```

You can also reset the positions completely by using the `set_pos_from_f` method. If you pass a seed, it will make the resulting positions the same every time.

```python
reeb_torus.set_pos_from_f(seed = 12)
reeb_torus.draw()
```

There are also some other example graphs, such as the dancing man

```python
dm = ex_reebgraphs.dancing_man()
dm.draw()
```

The next is an illustration of how loops are handled in the `plot_reeb()` function.  The `cpx` parameter controls how loopy the loops are.

```python
sl = ex_reebgraphs.simple_loops()
sl.draw(cpx=.1)
```

```python
sl.draw(cpx=1)
```

We can subdivide edges at certain function values.  To demonstrate this we will use the `reeb_torus` graph.

```python
R = ex_reebgraphs.torus()
R.draw_straight_line(with_labels = True)
```

Let's add a node  named `6` with function value 4.5 subdividing the edge `(4,5)`.

```python
R.subdivide_edge(u=4,v=5,w = 6, f_w=4.2)
R.draw_straight_line(with_labels = True)
```

We can also remove all regular nodes (i.e. those with both up- and down-degree equal to 1) to get an equivalent, minimal Reeb graph. Note that we're using the `draw` command here since `draw_straight_line` will make the multiedges land on top of each other and so not be visible.

```python
R.remove_all_regular_vertices()
R.draw(cpx = 15)
```
